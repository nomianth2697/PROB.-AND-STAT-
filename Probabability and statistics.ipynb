COMULATIVE DITRIBUTION FUNVTION(CDF)

```python
counts, bin_edges = np.histogram(iris_setosa['PetalLengthCm'],
bins=10,density=True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)

# compute CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.xlabel("PetalLength")
plt.ylabel("Provabablity")
plt.show()
```

BOX-COX Transformation

```python
from scipy import stats
import matplotlib.pyplot as plt
# if we have to convert form pareto distibution to gaussion distribution
```

```python
# before box cox transformation
fig = plt.figure()
ax1 = fig.add_subplot(211)
# this is our data we create data as x
x = stats.loggamma.rvs(5,size=500)+5
prob =  stats.probplot(x,dist=stats.norm,plot = ax1)
ax1.set_xlabel("")
ax1.set_title('Probplot afainst normal distribution / (pareto distiburation)')
plt.show()
```

```python
# after BOX-COX Transformation
fig = plt.figure()
ax2 = fig.add_subplot(211)
# this is our data we create data as x
x = stats.loggamma.rvs(5,size=500)+5
xt,_ = stats.boxcox(x)
prob = stats.probplot(xt,dist=stats.norm,plot=ax2)
ax2.set_xlabel("")
ax2.set_title('after box cox transformation (gaussion distribution) ')

plt.show()
```
